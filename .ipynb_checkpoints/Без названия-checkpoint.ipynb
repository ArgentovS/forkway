```python
import nltk
import re
import pandas as pd
import missingno as msno
from tqdm.auto import tqdm, trange
```

```python
# Загрузим датасет с исходными данными
df = pd.read_excel('2023-01-01-2023-01-10.xlsx')
```

```python
# Сформируем список столбцов исходного датасета
# с размеченными навыками для вакансии
cols = ['id']
for col in df.columns:
    if 'key' in col:
        cols.append(col)
```

```python
# Сформируем датафрейм размеченных навыков
df_skills = df[cols]
msno.matrix(df_skills);
```

```python
# Сформируем список уникальных навыков в исходных данных
unique_skills = []
for col in cols[1:]:
    unique_skills.extend(df_skills[col].unique())
unique_skills = [skill for skill in list(set(unique_skills))
                 if pd.notnull(skill)]
```

```python
# Сформируем датафрейм с векторами таргетов (признаками навков)
df_y = pd.concat([df['id'],
                 pd.DataFrame(columns=unique_skills)],
                 axis=1).fillna(0)
for i, row in df[cols[1:]].iterrows():
    for key, skill in row.iteritems():
        if pd.notnull(skill):
            df_y.loc[i, skill] = 1
```
