# 2. Препарирование данных для обучения

```python
# Загрузим датасет с текстом заявления и ключом для DWH
df = pd.read_excel('2023-01-01-2023-01-10.xlsx')
```

## 2.1. Преобразование исходных текстов

```python
# Удаляем из текста html-теги, неалфафитные символы и переводим в нижний регистр
df['normal_description'] = df.description.apply(lambda x:
                    re.sub('[^a-zа-я\+]', ' ', re.sub('<.+?>','', x).lower()))
# Лимматизируем слова и исключаем стоп-слова
df['normal_description'] = df['normal_description'].apply(lambda x:
                        text_converter(x))
```

## 2.2. Построение векторов таргетов

```python
# Сформируем список столбцов исходного датасета
# с размеченными навыками для вакансии
cols = ['id'] # в первом элементом оставим 'id' для проверки преобразований
for col in df.columns:
    if 'key' in col:
        cols.append(col)
# Сформируем и отобразим датафрейм размеченных навыков
df_skills = df[cols]
msno.matrix(df_skills);
```

```python
# Сформируем список уникальных навыков в исходных данных
unique_skills = []
for col in cols[1:]: # начинаем со второго столбца, т.к. в первом храниться id
    unique_skills.extend(df_skills[col].unique())
unique_skills = [skill for skill in list(set(unique_skills))
                 if pd.notnull(skill)]
```

```python
# Сформируем датафрейм с векторами таргетов (признаками навков)
df_y = pd.concat([df['id'],
                 pd.DataFrame(columns=unique_skills)],
                 axis=1).fillna(0)
for i, row in df[cols[1:]].iterrows():
    for key, skill in row.iteritems():
        if pd.notnull(skill):
            df_y.loc[i, skill] = 1
```

```python
df_y.drop(['id'], axis=1, inplace=True)
```

# 3. Обучение моделей

## 3.1. Формируем фичей и таргет

```python
# Формируем фичи и таргет
X = df['normal_description']
col = 'key_skills_name'
y = df_y[cols[:1]]#pd.read_excel('2023-01-01-2023-01-10.xlsx')[[col]].fillna('НЕТ')
```

```python
# Загружаем тренировочные и тестовые выборки
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                     random_state=42)
```

## 3.1. Модель на наивном классификаторе Баеса

```python
# Загружаем модели
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
```

```python
# Формируем пайплайн для последовательного преобразования данных в модели
nb = Pipeline([('vect', CountVectorizer()),
               ('tfidf', TfidfTransformer()),
               ('clf', MultinomialNB())])
```

```python
nb.fit(X_train, y_train)
```

```python
# Выполним тесирование
y_pred = nb.predict(X_test)
```

```python
# Построим метрики
from sklearn.metrics import classification_report, accuracy_score
print(f'accuracy: {round(accuracy_score(y_pred, y_test)*100, 2)}%')
#print(classification_report(y_test, y_pred, ))
```

## 3.2. Модель на машине опорных векторов

```python
# Загрузим модель, сформируем пайплайн, проведём обучение и тестирование
from sklearn.linear_model import SGDClassifier
sgd = Pipeline([('vect', CountVectorizer()),
               ('tfidf', TfidfTransformer()),
               ('clf', SGDClassifier())])
sgd.fit(X_train, y_train)
y_pred = sgd.predict(X_test)
```

```python
# Построим метрики
print(f'accuracy: {round(accuracy_score(y_pred, y_test)*100, 2)}%')
#print(classification_report(y_test, y_pred, ))
```

## 3.3. Модель логистической регрессии

```python
# Загрузим модель, сформируем пайплайн, проведём обучение и тестирование
from sklearn.linear_model import LogisticRegression
logreg = Pipeline([('vect', CountVectorizer()),
               ('tfidf', TfidfTransformer()),
               ('clf', SGDClassifier(n_jobs=1))])
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
# Построим метрики
print(f'accuracy: {round(accuracy_score(y_pred, y_test)*100, 2)}%')
#print(classification_report(y_test, y_pred, ))
```

```python
model_lstm = Sequential()
model_lstm.add(Embedding(num_words, 128, input_length=max_comment_len))
model_lstm.add(SpatialDropout1D(0.5))
model_lstm.add(LSTM(40, return_sequences=True))
model_lstm.add(LSTM(40))
model_lstm.add(Dense(6, activation='sigmoid'))
```
